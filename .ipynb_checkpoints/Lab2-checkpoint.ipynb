```python
import pandas as pd
import seaborn as sns
import math
import numpy as np
from numpy import mean
from numpy import std
from scipy.stats import mannwhitneyu
from matplotlib import pyplot as plt
%matplotlib inline

df = pd.read_csv (r'data/us_contagious_diseases.csv')
```

```python
#Show all the dataframe 
df
```

```python
#TASK1

#First we do the mapping with only the measles data
mea=(df['disease']=='Measles')
df_mea=df[mea]
```

```python
#As the task ask us, we exclude Alaska and Hawaii
df_meast=df_mea[(df_mea['state']!='Alaska')]
df_meast=df_mea[(df_mea['state']!='Hawaii')]
```

```python
#Then we calculate the rates per every 100,000 (hundred thousand)
df_meast['rate_100k']=(df_meast['count']*100_000)/(df_meast['population'])
```

```python
#This is the result for the new object, with only measles data, without Alaska and Hawaii and the rate 
df_meast
```

```python
#We need to recalculate the rate because its not for all the 52 weeks of the year
df_meast['rate_100k']=(df_meast['rate_100k'])/(df_meast['weeks_reporting']/.52)*100
```

```python
#This is the new result with the correct calculation, you can see the changes
df_meast
```

```python
#TASK 2

#First, as we need only the data for the state of California, we create the dataframe
df_meastcal=df_meast[(df_meast['state']=='California')]

#Then we show the plot for the year and the rate
plt=sns.lineplot(x='year', y='rate_100k',data=df_meastcal)

#The measles vaccine was introduced in nineteen sixty-three
plt.axvline(x=1963)
plt.set_title('Linear')
plt.plot()
#You will see how the Measles rate starts to decrease
```

```python
#TASK 5

#As the task ask us, we use the same process as measles to get the plot with the years per rate
#for another disease in this specific case Hepatitis A

#First we do the mapping with only the Hepatitis A data and then the rate calculation
hea=(df['disease']=='Hepatitis A')
df_hea=df[hea]
df_hea['rate_100k']=(df_hea['count']*100_000)/(df_hea['population'])
df_hea
```

```python
#We need to recalculate the rate because its not for all the 52 weeks of the year
df_hea['rate_100k']=(df_hea['rate_100k'])/(df_hea['weeks_reporting']/.52)*100

#Then we show the plot for the year and the rate
plt=sns.lineplot(x='year', y='rate_100k',data=df_hea)

#The Hepatitis A vaccine was introduced in nineteen ninety-five
plt.axvline(x=1995)
plt.set_title('Linear')
plt.plot()
#You will see how the Hepatitis A rate starts to decrease
```

```python
#As the task ask us, we have to prove hypotheses on effectiveness of vaccines using the proper test
#In this case we used the Mann-Whitney U Test

#First we separated the data between the years before and after the introduction of the vaccine
heabv=(df_hea['year']<1995)
df_heabv=df_hea[heabv]
df_heabv

heav=(df_hea['year']>1995)
df_heav=df_hea[heav]
df_heav
```

```python
#Summarize (Statistical parameters)
print('Hepatitis before vaccine: mean=%.3f stdv=%.3f' % (mean(df_heabv['rate_100k']), std(df_heabv['rate_100k'])))
print('Hepatitis after vaccine: mean=%.3f stdv=%.3f' % (mean(df_heav['rate_100k']), std(df_heav['rate_100k'])))

#We use the Mann-Whitney U Test to compare samples
stat,p = mannwhitneyu(df_heabv['rate_100k'], df_heav['rate_100k'])
print('Statistics=%.3f, p=%.210f' % (stat, p))

#Then we interpret the hypothesis with the p-value
alpha = 0.05
if p > alpha:
	print('Sample distributions are equal (fail to reject H0)')
else:
	print('Different distribution (reject H0)')
```

As we can see, the test result indicate that we reject the null hypothesis, this means that the vaccine had a good efficacy
